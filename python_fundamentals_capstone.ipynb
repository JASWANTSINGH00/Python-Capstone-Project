```
# Task 1: Create and save DataFrames
import pandas as pd  # performing task step

employee_data = {  # performing task step
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],  # performing task step
    'Name': ['John Alter', 'Alice Luxumberg', 'Tom Sabestine', 'Nina Adgra', 'Amy Johny'],  # performing task step
    'Gender': ['M', 'F', 'M', 'F', 'F'],  # performing task step
    'City': ['Paris', 'London', 'Berlin', 'Newyork', 'Madrid'],  # performing task step
    'Age': [25, 27, 29, 31, 30]  # performing task step
}  # performing task step
employee_df = pd.DataFrame(employee_data)  # performing task step
employee_df.to_csv('employee.csv', index=False)  # performing task step

seniority_data = {  # performing task step
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],  # performing task step
    'Designation Level': [2, 2, 3, 2, 3]  # performing task step
}  # performing task step
seniority_df = pd.DataFrame(seniority_data)  # performing task step
seniority_df.to_csv('seniority.csv', index=False)  # performing task step

project_data = {  # performing task step
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005', 'A002', 'A005', 'A003', 'A001', 'A003', 'A001', 'A004', 'A004', 'A005'],  # performing task step
    'Project': ['Project 1', 'Project 2', 'Project 3', 'Project 4', 'Project 5', 'Project 6', 'Project 7', 'Project 8',  # performing task step
                'Project 9', 'Project 10', 'Project 11', 'Project 12', 'Project 13', 'Project 14'],  # performing task step
    'Cost': [1002000, 2000000, 4500000, 5500000, None, 680000, 400000, 350000, None, 300000, 2000000, 1000000, 3000000, 200000],  # performing task step
    'Status': ['Finished', 'Ongoing', 'Finished', 'Ongoing', 'Finished', 'Failed', 'Finished', 'Failed',  # performing task step
               'Ongoing', 'Finished', 'Failed', 'Ongoing', 'Finished', 'Finished']  # performing task step
}  # performing task step
project_df = pd.DataFrame(project_data)  # performing task step
project_df.to_csv('project.csv', index=False)  # performing task step
```

```
# Task 2 to 10
import numpy as np  # performing task step

employee_df = pd.read_csv('employee.csv')  # performing task step
seniority_df = pd.read_csv('seniority.csv')  # performing task step
project_df = pd.read_csv('project.csv')  # performing task step

costs = project_df['Cost'].tolist()  # performing task step
for i in range(len(costs)):  # performing task step
    if pd.isna(costs[i]):  # performing task step
        prev_val = next((costs[j] for j in range(i - 1, -1, -1) if not pd.isna(costs[j])), None)  # performing task step
        next_val = next((costs[j] for j in range(i + 1, len(costs)) if not pd.isna(costs[j])), None)  # performing task step
        if prev_val is not None and next_val is not None:  # performing task step
            costs[i] = (prev_val + next_val) / 2  # performing task step
        elif prev_val is not None:  # performing task step
            costs[i] = prev_val  # performing task step
        elif next_val is not None:  # performing task step
            costs[i] = next_val  # performing task step
project_df['Cost'] = costs  # performing task step

employee_df[['First Name', 'Last Name']] = employee_df['Name'].str.split(' ', 1, expand=True)  # performing task step
employee_df.drop(columns=['Name'], inplace=True)  # performing task step

merged1 = pd.merge(employee_df, seniority_df, on='ID')  # performing task step
final_df = pd.merge(merged1, project_df, on='ID')  # performing task step

final_df['Bonus'] = final_df.apply(lambda row: 0.05 * row['Cost'] if row['Status'] == 'Finished' else 0, axis=1)  # performing task step

final_df.loc[final_df['Status'] == 'Failed', 'Designation Level'] += 1  # performing task step
final_df = final_df[final_df['Designation Level'] <= 4]  # performing task step

final_df['First Name'] = final_df.apply(lambda row: 'Mr. ' + row['First Name'] if row['Gender'] == 'M' else 'Mrs. ' + row['First Name'], axis=1)  # performing task step
final_df.drop(columns=['Gender'], inplace=True)  # performing task step

final_df.loc[final_df['Age'] > 29, 'Designation Level'] -= 1  # performing task step
final_df['Designation Level'] = final_df['Designation Level'].clip(lower=1)  # performing task step

total_proj_cost = final_df.groupby(['ID', 'First Name'])['Cost'].sum().reset_index()  # performing task step
total_proj_cost.columns = ['ID', 'First Name', 'Total Cost']  # performing task step

city_with_o = final_df[final_df['City'].str.contains('o', case=False)]  # performing task step

final_df.to_csv('final_df.csv', index=False)  # performing task step
total_proj_cost.to_csv('total_proj_cost.csv', index=False)  # performing task step
city_with_o.to_csv('city_with_o.csv', index=False)  # performing task step

final_df.head()  # performing task step
```
